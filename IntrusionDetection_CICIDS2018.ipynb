```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif, chi2
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Drop unnecessary columns
data = df.drop(['Label', 'Attack Type', 'Timestamp'], axis=1)

# Separate features and target
X = data.drop('Attack Type Code', axis=1)
y = data['Attack Type Code']
```

```python

# Remove constant features
constant_features = [col for col in X.columns if X[col].nunique() <= 1]
```

```python
df[constant_features]
```

```python
X = X.drop(columns=constant_features) #drop constant features
```

```python
# Define numerical preprocessor
numerical_preprocessor = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Define feature selector
select_k_best = SelectKBest(score_func=f_classif, k=30)  # Adjust k based on your requirement

# Apply feature selection
X_selected = select_k_best.fit_transform(X, y)
```

```python
selected_indices = select_k_best.get_support(indices=True)
selected_feature_names = X.columns[selected_indices]

# Ensure X is a DataFrame before using ColumnTransformer
X_selected_df = pd.DataFrame(X_selected, columns=selected_feature_names)

# Define column transformer for the selected features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_preprocessor, selected_feature_names)
    ]
)

# Apply scaling to the selected features
X_preprocessed = preprocessor.fit_transform(X_selected_df)
```

```python
# Save the scaler and feature selector
import joblib
scaler = preprocessor.named_transformers_['num'].named_steps['scaler']
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(select_k_best, 'feature_selector.pkl')
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
import pandas as pd

# Apply less aggressive undersampling to retain enough samples for SMOTE
undersample = RandomUnderSampler(sampling_strategy={
    0: 1000000,  # Keep 1,000,000 samples of the majority class 'Benign'
    3: 300000,   # Keep 300,000 samples of 'DDoS'
    4: 300000    # Keep 300,000 samples of 'DoS'
}, random_state=42)

# Apply undersampling first to the dataset
X_undersampled, y_undersampled = undersample.fit_resample(X_preprocessed, y)

# Adjust the SMOTE sampling strategy to avoid oversampling more than the existing samples
smote = SMOTE(sampling_strategy={
    6: 50000,   # SQL Injection (from 87 to 50,000)
    5: 161060,  # Infiltration (up to the original count)
    2: 157509,  # Brute Force (up to the original count)
    1: 290000,  # Bot (up to 200,000)
    4: 300000,  # DoS (Retain the same size after SMOTE)
    3: 300000   # DDoS (Retain the same size after SMOTE)
}, random_state=42)

# Apply SMOTE to oversample the minority classes
X_resampled, y_resampled = smote.fit_resample(X_undersampled, y_undersampled)

# Check the new class distribution
print(pd.Series(y_resampled).value_counts())
```

```python
X_resampled.shape
```

```python
# Train RandomForestClassifier to get feature importances
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_resampled, y_resampled)
importances = clf.feature_importances_
```

```python
# Visualize feature importances
plt.figure(figsize=(12, 8))
plt.title('Feature Importances')
sns.barplot(x=importances, y=selected_feature_names, palette="viridis")
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score, mean_squared_error
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from imblearn.pipeline import Pipeline as ImbalancedPipeline
```

## Logic to Evaluate Models

```python
def eval_metric(pipeline, X_train, y_train, X_test, y_test):
    # Predicting on the training set
    y_train_pred = pipeline.predict(X_train)
    # Predicting on the test set
    y_pred = pipeline.predict(X_test)
    
    print("\nTest Set Evaluation")
    print("===================")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    print("\nTrain Set Evaluation")
    print("====================")
    print(f"Accuracy: {accuracy_score(y_train, y_train_pred):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_train, y_train_pred))
    print("\nClassification Report:")
    print(classification_report(y_train, y_train_pred))
```

```python
# Define parameter grids for models
param_grids = {
    'GaussianNB': {},
    'RandomForestClassifier': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'XGBClassifier': {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 6, 9],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.8, 1.0]
    }, 
    'LogisticRegression': {
        'C': [0.1, 1, 10],
        'penalty': ['l2'],
        'solver': ['liblinear', 'saga']
    },
    'SVC': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    },
}

# Initialize best model tracker
best_model = None
best_score = 0
best_model_name = ''

# Define models
models = {
    'GaussianNB': GaussianNB(),
    'RandomForestClassifier': RandomForestClassifier(random_state=42),
    'XGBClassifier': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    'SVC': SVC(probability=True, random_state=42),
    'LogisticRegression': LogisticRegression(max_iter=1000, random_state=42)
}


# Grid Search for hyperparameter tuning
for model_name, model in models.items():
    print(f"\nTuning {model_name}...")

    
    param_grid = param_grids[model_name]
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy', n_jobs=-1, verbose=1)
    grid_search.fit(X_resampled, y_resampled)
    
    # Best parameters and score
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    print(f"Best score for {model_name}: {grid_search.best_score_}")
    
    # Save the best model
    joblib.dump(grid_search.best_estimator_, f'models/{model_name}_best.pkl')
    
    # Evaluate the model on the training and test sets
    print(f"\nEvaluating {model_name}...")
    eval_metric(grid_search.best_estimator_, X_resampled, y_resampled, X_resampled, y_resampled)
    
    # Update best model
    if grid_search.best_score_ > best_score:
        best_score = grid_search.best_score_
        best_model = grid_search.best_estimator_
        best_model_name = model_name

print(f"\nBest model: {best_model_name} with score: {best_score}")
```
